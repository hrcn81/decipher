```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/hrcn81/commonplace/main/Datasets/pizza_v1.csv')
```

```python
df = data.copy()
```

```python
df.head()
```

```python
df.shape
```

```python
x = 1
for i in df["topping"].unique():
    print(x,i)
    x += 1
```

```python
df['topping'].nunique()
```

```python
df['price_rupiah'] = df['price_rupiah'].str.replace('Rp','').str.replace(',','.')
```

```python
df.head(2)
```

```python
# df["topping"] = df["topping"].replace({"chicken": 1})

df['extra_sauce'] = df['extra_sauce'].replace({'yes' : 1,'no':0})
df['extra_cheese'] = df['extra_cheese'].replace({'yes' : 1,'no':0})
```

```python
df['price_rupiah'] = df['price_rupiah'].astype(float)
df['diameter'] = df['diameter'].astype(float)
```

```python
df.info()
```

```python
df.rename(columns = {'price_rupiah' : 'price'},inplace = True)
```

```python
df.head()
```

```python
df['company'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(df['company'])

df['company'] = le.transform(df['company'])
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/hrcn81/commonplace/main/Datasets/scholarship.csv')
```

```python
df = data.copy()
```

```python
df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.subplot(1,2,1)
sns.distplot(df['semester_percentage'])

plt.subplot(1,2,2)
sns.distplot(df['scholarship_exam_marks'])

plt.show()
```

```python
mean = df['semester_percentage'].mean()

std = df['semester_percentage'].std()
```

```python
print(mean,' ',std)
```

```python
highest_allowed = mean + 3 * std

lowest_allowed = mean - 3 * std
```

```python
highest_allowed
```

```python
lowest_allowed
```

```python
df.shape
```

# Trimming

```python
df[(df['semester_percentage'] < lowest_allowed) | (df['semester_percentage'] > highest_allowed)].shape
```

```python
df[(df['semester_percentage'] > lowest_allowed) & (df['semester_percentage'] < highest_allowed)].shape
```

```python
df['z_score'] = (df['scholarship_exam_marks'] - df['semester_percentage'].mean())/df['semester_percentage'].std()
```

```python
df['z_score']
```

```python
cols = 0
for i in data.columns:
    cols += 1
    print(cols,i)
```

```python
cols = 0
for i in df.columns:
    cols += 1
    print(cols,i)
```

```python
df.head()
```

```python
df[(df['z_score'] < -3) | (df['z_score'] > 3)]
```

```python
df[(df['z_score'] > -3) & (df['z_score'] < 3)]
```

# Capping

```python
upper_limit = mean + 3 * std
lower_limit = mean - 3 * std
```

```python
import numpy as np
```

```python
df['semester_percentage'] = np.where(df['semester_percentage'] > upper_limit,upper_limit,
                                    np.where(df['semester_percentage'] < lower_limit,lower_limit,
                                             df['semester_percentage']) )
```

```python
df
```

```python
df[(df['semester_percentage'] > lowest_allowed) & (df['semester_percentage'] < highest_allowed)].shape
```

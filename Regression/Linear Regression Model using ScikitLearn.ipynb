```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Algos/data_for_lr.csv')
```

```python
data = data.dropna()
```

```python
data.shape
```

```python
train_ip = np.array(data.x[:500]).reshape(500,-1)
train_op = np.array(data.y[:500]).reshape(500,-1)

test_ip = np.array(data.x[500:700]).reshape(199,-1)
test_op = np.array(data.y[500:700]).reshape(199,-1)
```

# Linear Regression

## Training Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(train_ip,train_op)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

## Prediction

```python
test_predictions = lr_model.predict(test_ip)
```

```python
plt.plot(test_ip,test_predictions,'+',color = 'red')
plt.plot(test_ip,test_op,'.',color = 'green')
plt.title('Performance Testing')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
# plt.plot
```

## Cost Function

```python
from sklearn.metrics import mean_squared_error
```

```python
cost = mean_squared_error(test_op,test_predictions)
```

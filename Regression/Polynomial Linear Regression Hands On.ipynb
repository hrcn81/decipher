```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading Data

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Algos/Position_Salaries.csv')
```

```python
df = data.copy()
```

# Splitting the data

```python
x = df.iloc[:,1].values.reshape(10,-1)
y = df.iloc[:,-1].values.reshape(10,-1)
```

```python
x
```

```python
y
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()

lin_reg.fit(x,y)
```

# Visualize the LR Model

```python
plt.scatter(x,y,color = 'red')
plt.plot(x,lin_reg.predict(x),color = 'blue')
plt.xlabel('input')
plt.ylabel('prediction')
plt.show()
```

# Polynomial linear Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree = 2)

x_poly = poly_reg.fit_transform(x)
```

```python
x_poly
```

```python
poly_linear_reg = LinearRegression()

poly_linear_reg.fit(x_poly,y)
```

```python
plt.scatter(x,y,color = 'red')
plt.plot(x,poly_linear_reg.predict(poly_reg.fit_transform(x)),color = 'blue')
plt.xlabel('input')
plt.ylabel('prediction')
plt.show()
```

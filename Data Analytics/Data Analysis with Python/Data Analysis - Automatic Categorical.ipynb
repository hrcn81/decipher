```python
ART_AND_DESIGN = 0
for i in dcat:
  if i == 'ART_AND_DESIGN':
    ART_AND_DESIGN += 1

ART_AND_DESIGN
```

```python
AUTO_AND_VEHICLES = 0
for i in dcat:
  if i == 'AUTO_AND_VEHICLES':
    AUTO_AND_VEHICLES += 1

AUTO_AND_VEHICLES
```

```python
COMICS = 0
for i in dcat:
  if i == 'COMICS':
    COMICS += 1

COMICS
```

```python
for name in dcat.unique():
  ct = 0
  for i in dcat:
    if i == name:
      ct += 1
  print(name,ct)
```

```python
for name in dcat.unique():
    ct = 0
    for i in dcat:
        if i == name:
          ct += 1
    print(f'{name} : {ct}')
```

```python
categories = {}

for name in dcat.unique():
    ct = 0
    for i in dcat:
        if i == name:
          ct += 1
    categories[name] = ct
    # print(f'{name} : {ct}')

categories
```

```python
# Assuming 'dcat' is your Series containing content categories
category_counts = pd.DataFrame(index=dcat.unique(), columns=['Count'])

for name in dcat.unique():
    ct = len(dcat[dcat == name])
    category_counts.loc[name, 'Count'] = ct

# Displaying the DataFrame
print(category_counts)
```

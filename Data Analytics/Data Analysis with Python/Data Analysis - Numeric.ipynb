# Q.2 | How Many apps are there with rating 5 ?

```python
df["Rating"]
```

```python
c = 0
for i in df['Rating']:
  if (i == 5.0):
    c += 1
# print(c)
print(f'There are {c} Apps with rating 5.')
```

# Q.3 | How Many apps are there with rating 4.0 to 4.5 ?

```python
d = 0
for i in df['Rating']:
  if (i >= 4.0 and i <= 4.5):
    d += 1
# print(d)
print(f'There are {d} Apps with rating between 4.0 to 4.5.')
```

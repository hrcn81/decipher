# 2. Imputers

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
```

```python
df.iloc[:,1:3].values
```

```python
imputer.fit(df.iloc[:,1:3].values)

x = imputer.transform(df.iloc[:,1:3].values)

x
```

```python
df.iloc[:,1:3] = x

df
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/pandas/Data.csv')

df
```

```python
imputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')

imputer.fit(df.iloc[:,1:2].values)

df.iloc[:,1] = imputer.transform(df.iloc[:,1:2].values)
```

```python
df.iloc[:,1:2]
```

```python
imputer.fit(df.iloc[:,2:3].values)

df.iloc[:,2] = imputer.transform(df.iloc[:,2:3].values)
```

```python
df.iloc[:,2:3]
```

```python
df
```

```python
imputer.fit(df.iloc[:,:1].values)

df.iloc[:,0] = imputer.transform(df.iloc[:,:1].values)
```

```python
df
```

```python
imputer.fit(df.iloc[:,-1:].values)

df.iloc[:,-1] = imputer.transform(df.iloc[:,-1:].values)
```

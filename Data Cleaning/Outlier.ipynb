```python
dfo2 = pd.DataFrame(dataout,columns = ['Col_1'])

dfn2 = pd.DataFrame(datanotout,columns = ['Col_1'])
```

```python
dfo2.mean()
```

```python
dfo2.median()
```

```python
dfo2.mode()
```

```python
dfo2.describe()
```

```python
# With Outlier

dfo2.agg(['mean', 'median', lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else np.nan, 'var', 'std'])
```

```python
# Without Outlier

dfn2.agg(['mean', 'median', lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else np.nan, 'var', 'std'])
```

```python
cities = ['New york','Los Angeles','Chicago','Houston','Phoenix','Los Angeles','Chicago']
rankings = [1,2,3,4,5,6,7]
scores = [9.8,9.5,9.2,8.7,8.5,8.2,8.1]

data = {'City':cities,'Ranking': rankings,'Score':scores}

dfd = pd.DataFrame(data)
```

```python
print(dfd.drop_duplicates('City',keep = False),'\n')      # Removes all duplicate
print(dfd.drop_duplicates('City',keep = 'first'),'\n')    # by default removes all duplicate and keep first value of it
print(dfd.drop_duplicates('City',keep = 'last'),'\n')     # Remove duplicate and keep last values of it
```
